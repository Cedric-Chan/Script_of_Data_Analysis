# 1. 读取数据并插入数据库

```python
# 创建数据库并把 txt文件的数据存进数据库

import sqlite3      #导入sqlite3
import os
path = os.getcwd()
files = os.listdir(path)
```

## 1.1数据库初始化

```python
conn = sqlite3.connect(path+'\db.db') # path是程序所在路径，也是TXT文件的所在路径
# 如果数据库已经存在，则连接数据库；如果数据库不存在，则先创建数据库，再连接该数据库。

cur = conn.cursor()           # 创建游标，以便获得查询对象。

cur.execute('create table if not exists numbers (id integer primary key ,number varchar(20) NOT NULL)') 
# 在数据库db.db 中创建表numbers,表中包含两个字段：id、number。

conn.commit()
# 不要忘了conn.commit()保存一下！不然后面的数据是插不进去的。
```

## 1.2写入数据

```python
# 插入数据

i = 0
for file in files:
    if file.split('.')[-1] == 'txt':

        # 使用 with 语句来创建文件对象，对象会在with语句结束时自动关闭。效率更高！不易出错。
        with open(file, 'r', encoding='UTF-8') as f:
            next(f)
            for line in f:
                i += 1
                print("插入第", i, "条数据:")
                print(line)
                cur.execute('insert into numbers values(?,?)', (i, line))
                # (?,?) 起到占位符功能，(i,line))中的值会按位置替换到SQL语句中
                conn.commit()
```

## 1.3关闭服务器，保存

```python
cur.close() # 关闭游标
conn.close() # 关闭数据库连接
print('数据写入完成！共写入',i,' 条数据')
```

# 2.数据库的操作

## 2.1连接数据库

```python
import sqlite3
import os
path = os.getcwd()
print(path)
files = os.listdir(path)
conn = sqlite3.connect(path+'\db.db') # path+'\db.db'是数据库的路径及文件名，也支持手动输入路径
cur = conn.cursor()
```

## 2.2查询数据库中有哪些表

```python
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
Tables=cur.fetchall()
print(Tables)
```

## 2.3删除数据库中的某个表

```python
cur.execute("drop table numbers;")
```

## 2.4查询某个表的结构

```python
cur.execute("PRAGMA table_info(numbers)")
print(cur.fetchall())
```

## 2.5查询表中前50条记录

```python
cur.execute("SELECT * from numbers limit 0,50;")
conn.commit
data = cur.fetchall()
print(data)
```

## 2.6查询表中所有记录

```python
cur.execute("SELECT * from numbers;")
data_all = cur.fetchall()
a = len(data_all)
print('共有 '+ str(a) + ' 条记录')
#print(data_all)
```

## 2.7查询表中不重复记录

```python
cur.execute("SELECT distinct number from numbers;")
data_distinct = cur.fetchall()
b = len(data_distinct)
print('共有 '+ str(b) +' 条不重复记录')
#print(data_distinct)
```

## 2.8将旧表中的不重复记录插入新表

```python
# 创建一个新表numbers_distinct，包含id和number两个字段
cur.execute('create table if not exists numbers_distinct (id integer primary key ,number varchar(20) NOT NULL)')
conn.commit()
# 插入数据
i = 0
for data in data_distinct:
    i += 1
    data = data[0]
    #print(data)
    cur.execute('insert into numbers_distinct values(?,?)',(i,data))
    conn.commit()
```

## 2.9将特定结果写入文本文件（单列）

```python
cur.execute("SELECT number from numbers_distinct limit 0,10;")  # 提取前十条
datas = cur.fetchall()
print(datas)

with open('datafile1.txt','w') as f1:
    for data in datas:
        f1.write(data[0])
        f1.flush()
        # f1.flush()操作很重要！将缓冲区的数据写入文件中，否则文本文件为空白，导致写入失败。
```

## 2.10将特定结果写入文本文件（多列）


```python
cur.execute("SELECT * from numbers_distinct limit 0,10;")
datas = cur.fetchall()
print(datas)
with open('datafile2.txt','w') as f2:
    for data in datas:
        data0 = str(data[0]) # 将int类型转为str，否则write函数报错
        data = data0 + '    ' + data[1] # 含索引列，在两列之间以Tab键分隔
        f2.write(data)
        f2.flush()
```

## 2.11将特定结果写入CSV文件（多列）

```python
with open('datafile3.csv', 'w') as f3:      #文件后缀改为 '.csv'
    for data in datas:
        data0 = str(data[0])  # 将int类型转为str，否则write函数报错
        data = data0 + ',' + data[1]  # 在两列之间以逗号键（半角字符）分隔
        f3.write(data)  # data是元组类型
        f3.flush()  # 重要！将缓冲区的数据写入文件中
```

# 3.办公自动化案例


#### 将number_distingct表中的数据，以每50个为一组，分别存入一个文本文件

```python
b = len(data_distinct)

for n in range(int(b / 50) + 1):
    if 50 * n + 50 < b:  # b为数据记录总数
        cur.execute('SELECT number from numbers_distinct limit ?,?',(50 * n, 50))
        datas = cur.fetchall()
        # print(datas)
        filename = str(
            50 * n) + '-' + str(50 * n + 50) + '.txt'  # 重点是如何计算出每个TXT文件的名称
        print('正在写入：' + filename)
        
        with open(filename, 'w') as f4:
            for data in datas:
                f4.write(data[0])  # data是元组类型
                f4.flush()  # 重要！将缓冲区的数据写入文件中
                
    else:
        filename = str(50 * n) + '-' + str(b) + '.txt'  # 最后一个文件名的构造
        print('正在写入：' + filename)
        with open(filename, 'w') as f5:  # 计算最后一个文本文件名称
            cur.execute('SELECT number from numbers_distinct limit ?,?',
                        (50 * n + 50, 50))
            datas = cur.fetchall()
            for data in datas:
                f5.write(data[0])  # data是元组类型
                f5.flush()  # 重要！将缓冲区的数据写入文件中
                
print('写入完成！共写入{}个TXT文件。'.format(n + 1))
```
