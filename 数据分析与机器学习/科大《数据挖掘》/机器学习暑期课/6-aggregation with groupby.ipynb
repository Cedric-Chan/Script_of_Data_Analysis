```python
# # 不同job，求知道SQL的学生占比
df.groupby('Job')['SQL'].mean()   # 二级groupby，同df.groupby('Job').SQL.mean() 
df.groupby('Program')['SQL'].sum()   # sum包含全部，size只计算1的值

#! 对于会SQL的学生，求每个program中会Java的学生比例
sql = df[df['SQL']==1]
sql.groupby('Program')['Java'].mean()

# 对每个classification，找到有多少MBA学生？
df[df.Program=='MBA'].groupby('Classification').size()
```

##### mutiple functions(agg)--显示多个求值结果，以dataframe显示

```python
df.groupby('Job')['SQL'].agg(['mean','size']).rename(columns={'mean':'SQL_prop','size':'n_students'})
```

```python
result = df.groupby('Job').agg({'SQL':'mean','Classification':['max',lambda x: x.max()-x.min()]})
result.columns = ['mean_SQL','max_Class','spread_Class']
result
```

problem

```python
# find max，min, average of Languages known by students in each Program
df.groupby('Program')['Languages'].agg(['max','min','mean'])

# 对于programming level 和 Program，计算知道Python的人数和比例
df.groupby(['Program','ProgSkills']).agg({'Job':'size','Python':'mean'}).rename(columns={'Job':'nStudents','Python':'Python_prop'})

# 对于每个program，计算同时会Python和C的人数；Clustering中最大最小值差值
df2=df.copy()
df2['P&C']=df2['Python']*df2['C']==1
df2.groupby('Program').agg({'P&C':'sum','Clustering':lambda x:x.max()-x.min()})
```

##### DataFrame  Groupby()
更灵活，但更慢

<b>groupby()之后,lambda()提取部分row</b>  

```python
#  对于不同的program，找到知道语言最多的学生
df.groupby('Program')['Languages'].max()  # 各专业最多语言数
df[df.Languages == df.Languages.max()]    # 知道最多语言数的学生，不分专业

## 将大表拆成所有专业的小表，然后进行第二个代码
df.groupby('Program').apply(lambda d: d[d.Languages == d.Languages.max()])
```

```python
df.groupby('Program')['ProgSkills'].mean()  # 速度更快的方法
df.groupby('Program').apply(lambda d: d.ProgSkills.mean())  # 更灵活的方法
```

<b>Problem1</b>

```python
# for each ProgSkill, find students with highest Classification and show their C & Java
df[df.Classification == df.Classification.max()]  # 生成boolean mask
df.loc[df.Classification == df.Classification.max(),['C','Java']]  # 只看C和Java

df.groupby('ProgSkills').apply(lambda d: d.loc[d.Classification == d.Classification.max(),['C','Java']])
```

```python
# 另解
# transform()生成所在学生所在progskill最大classification
df2=df.copy()
df2['max_classif_within_progskills']=df2.groupby('ProgSkills')['Classification'].transform('max')
df2.head()
```

<b>Problem2</b>

```python
# for each ProSkills, find Program with most students that have that progskill level
df.groupby('Program').size().nlargest(1)  # 先找最多学生的program

df.groupby('ProgSkills').apply(lambda d: d.groupby('Program').size().nlargest(1))
```

##### transform()生成所在类别{最大、平均……}某个字段参数

create a 0-1 column whose value is 1 if the student konws more languages than the average of his program

```python
df2=df.copy()
df2['avg_languages']=(df2.Languages > df2.groupby('Program')['Languages'].transform('mean'))*1
df2.head()
```

create a 0-1 column that whether a student knows clustering better than the average in his quarter

```python
df2=df.copy()
df2['avg_clustering']=(df2.Clustering > df2.groupby(['quarter','year'])['Clustering'].transform('mean'))*1
df2.head()
```
