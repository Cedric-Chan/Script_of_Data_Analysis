```python
df = pd.read_csv('cleaned_survey.csv',index_col=0)
df.head()
```

```python
df_programs = pd.DataFrame({
    'Program': [
        'MSIS', 'MBA', 'Master of Finance', 'Supply Chain Mgmt & Analytics',
        'Master of Hacking'
    ],
    'Units_required': [51, 70, 48, 49, 100]
})
```

#### merge on columns

##### inner merge(default)求交集

```python
df.merge(df_programs)  # 用唯一共有的column合并
```

##### left_on & right_on

```python
df_inner = df.merge(df_programs,left_on='Program',right_on='Program')
len(df_inner)
```

##### left merge()首先出现的dataframe

```python
df.merge(df_programs,how='left')  # 和左边dataframe一样的set
```

##### right merge()

```python
df.merge(df_programs,how='right')  # 和右边dataframe一样的set
```

#### merge on indices

```python
df_programs_i = df_programs.set_index('Program')
```

```python
data = df.merge(df_programs_i,left_on='Program',right_index=True)  # 只有一个index，how默认是inner
```

problems

```python
# for each programming skills level, find the average number of units to be completed by students with that programming skill levell
data.groupby('ProgSkills')['Units_required'].mean()

# for each existing program, find the units required to complete it and the number of students belonging to that program that responded to the survey
df.merge(df_programs,how='right').groupby('Program').agg({'Units_required':'mean','Program':'count'}).rename(columns={'Units_required':'Units_required','Program':'students_size'})

# for each person in df, the number fo weekly hours of working. assuming that:
#1 each required unit of coursework is 0.25 hours a week
#2 job=0 is 0 hour
#3 job = 0.5 is 20 hours a week
#4 job = 1 is 40 hours a week
df_left = df.merge(df_programs,how='left')
df_left['hours_per_week']=(df_left['Units_required']*0.25) +(df_left['Job']==0.5)*20 + (df_left['Job']==1.0)*40
df_left.head()
```

#### deal with Nan

```python
df_left.fillna(0,inplace=True)
df_left.isna().any()
```
